# Réaffectation des individus

Dans cette partie, l'objectif est de réaffecter les individus dans les groupes avec des variables illustratives. On commencera par affecter la segmentation variable en vert en utilisant les variables oranges, puis nous utiliserons un jeu de variables spécifique détaillé dans sa partie dédiée.

## Setup

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV, train_test_split 
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Classification en utilisant les variables oranges

```python
green = pd.read_csv("../data/vert.csv")
green
```

```python
orange = pd.read_csv("../data/orange.csv")
orange
```

```python
data = pd.merge(left=orange.iloc[:, :-1], right=green[["cluster", "Respondent_ID"]], how="right", left_on="Respondent_ID", right_on="Respondent_ID")
data = data.drop(columns=["Respondent_ID"])
data
```

```python
for col in data.columns:
    print(col, ":", data[col].unique())
```

```python
data = data.fillna(-1)
```

```python
data.isnull().values.any()
```

```python
# Représentation en 3D dans l'espace des composantes principales
pca = PCA(n_components=3)
features_pca = pca.fit_transform(data.values)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for cluster in range(4):
    cluster_indices = data[data['cluster'] == cluster].index
    ax.scatter(features_pca[cluster_indices, 0], features_pca[cluster_indices, 1], features_pca[cluster_indices, 2], label=f'Cluster {cluster}')
    
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.legend()
plt.show()
```

## Initialisation du modèle

Dans le cadre de cet exercice, nous nous concentrerons d'abord sur l'utilisation de l'agorithme Random Forest pour tenter de résoudre la tâche de classification avec un score élevé. Ce choix se justfie par les performances très élevées du Random Forest dans la réalisation de la Partie 2.

```python
X = data.drop(columns=['cluster'])
y = data['cluster']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)

print(len(X_train), len(X_test))
```

```python
n_estimators = 1000

rf_model = RandomForestClassifier(n_estimators=n_estimators, random_state=23)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

On se rend compte que le Random Forest associé à l'hyperparamètre utilisé lors de la Partie 2 n'est pas très efficace. En effet, le modèle n'obtient qu'un score de 50% pour l'accuracy. On va donc tenter d'optimiser l'hyperparamètre en utilisant la technique de recherche par grille.

```python
param_grid = {
    'n_estimators': list(range(1100, 2100, 100)),
    'max_depth': list(range(10, 60, 10)),
}
```

```python
# On configure le GridSearchCV
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=2,
    verbose=3,
    n_jobs=1,
)

# On lance la recherche
grid_search.fit(X_train, y_train)

# On affiche les meilleurs hyperparamètres
print("Meilleurs hyperparamètres :", grid_search.best_params_)
```

```python
# Récupérer le meilleur modèle
best_rf_model = grid_search.best_estimator_

# Évaluer sur les données de test
test_score = best_rf_model.score(X_test, y_test)
print(f"Score du modèle optimisé sur les données de test : {test_score:.2%}")
```

Malgré les optimisations, le modèle optimisé ne produit pas de meilleurs résultats et reste aux alentours de 49-50%.

```python
variables = variables.fillna(-1)
```

```python
data.isnull().values.any()
```

```python
data = pd.merge(left=variables, right=green[["cluster", "Respondent_ID"]], left_on="Respondent_ID", right_on="Respondent_ID", how="right")
data = data.drop(columns=["Respondent_ID"])

data
```

Etant donné que la colonne RS2 (l'âge du sondé) représente une valeur non catégorique (il y autant de réponses que d'âges indiqués par les sondés), on peut standardiser cette colonne.

```python
rs2_mean = np.mean(data["RS2"])
rs2_std = np.std(data["RS2"])

data["RS2"] = (data["RS2"] - rs2_mean) / rs2_std
data
```

```python
# Représentation en 3D dans l'espace des composantes principales
pca = PCA(n_components=3)
features_pca = pca.fit_transform(data.values)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for cluster in range(4):
    cluster_indices = data[data['cluster'] == cluster].index
    ax.scatter(features_pca[cluster_indices, 0], features_pca[cluster_indices, 1], features_pca[cluster_indices, 2], label=f'Cluster {cluster}')
    
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.legend()
plt.show()
```

```python
X = data.drop(columns=['cluster'])
y = data['cluster']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)

print(len(X_train), len(X_test))
```

```python
n_estimators = 1000

rf_model = RandomForestClassifier(n_estimators=n_estimators, random_state=23)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

L'accuracy est encore faible ici (seulement 44%). On va donc commencer par optimiser de nouveau les hyperparamètres du modèle.

```python
param_grid = {
    'n_estimators': list(range(1100, 2100, 100)),
    'max_depth': list(range(10, 60, 10)),
}
```

```python
# On configure le GridSearchCV
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=2,
    verbose=3,
    n_jobs=1,
)

# On lance la recherche
grid_search.fit(X_train, y_train)

# On affiche les meilleurs hyperparamètres
print("Meilleurs hyperparamètres :", grid_search.best_params_)
```

```python
# Récupérer le meilleur modèle
best_rf_model = grid_search.best_estimator_

# Évaluer sur les données de test
test_score = best_rf_model.score(X_test, y_test)
print(f"Score du modèle optimisé sur les données de test : {test_score:.2%}")
```

on va envisager d'utiliser un autre algorithme comme KNN par exemple.
